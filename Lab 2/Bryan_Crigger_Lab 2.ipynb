```python
## Creating the test, training, and validation datasets &
## Aggregating data by zip code/RegionName
drop_clms = ['SizeRank', 'State', 'City', 'Metro', 'CountyName']

zillow_train = zillowDF.drop(zillowDF.iloc[:,-27:], axis = 1)
zillow_train = zillow_train.drop(columns=drop_clms)
zillow_train = zillow_train.sort_values('RegionName')
zillow_train = zillow_train.set_index('RegionName').transpose()##.reset_index()
# zillow_train.index = pd.to_datetime(zillow_train['RegionName'])
# zillow_train = zillow_train.rename(columns = {'index':'ds'})
zillow_train
```

```python
## Adjusted function to loop through multiple prophet forecasts 
def beProphet(label,components,predPeriods=12):
    model={}
    ## restructure the dataframe to fit prophet
    df = pd.DataFrame(components)
    df = df.reset_index()
    logValue = 'log_' + label
    df[logValue] = np.log(df[label])
    dfProphet = df.rename(index=str, columns={label:'y','index':'ds'})
    dfProphet = dfProphet.loc[:,['y','ds']]
    
    ## setting confidence interval to 95%
    zipModel = Prophet(interval_width=0.95)
    zipModel_fit = zipModel.fit(dfProphet)
    model['model_fit'] = zipModel_fit
    
    ## make future dates dataframe
    future_dates = zipModel.make_future_dataframe(periods=predPeriods, freq='M', include_history=True)
    
    ## model
    forecast = zipModel.predict(future_dates)
    model['model_forecast'] = forecast
    
   
    return(model)
```

```python
# perform ZipCod model creation and validation techniques
# build time series models
# perform exploratory data analysis techiques
# Build prophet timeseries models for the metro area, save to dictionary object
zip_forecasts = {}
for RegionName, price in zillow_train.items():
    model = beProphet(RegionName,price,predPeriods=13)
    model = zip_forecasts[RegionName]
```
